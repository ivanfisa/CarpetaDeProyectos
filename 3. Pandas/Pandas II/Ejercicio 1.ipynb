### DATA FRAME DE LA URL

```python
#importamos librerias
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

#definimos la url y la variable para el dataframe
url = 'https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true'
df = pd.read_csv(url, index_col=0)
df.head(100000)
```

```python
df.info()
```

### 1. QUE PAÍSES SE ENCUENTRAN EN EL DATAFRAME

```python
tab0 = pd.crosstab(index=df['location'], columns='date')
tab0.head(10000)
```

### 2. CUANTOS REGISTROS HAY POR CONTINENTE

```python
tab = pd.crosstab(index=df['continent'], columns='location')
tab.head(10000)
```

### 3. CUANTOS NUEVOS CASOS HUBO COMO MAXIMO EN UNA FECHA REPORTADA POR CADA PAÍS

```python
df[df.date == '2021-05-29'].sort_values('new_cases', ascending=False)
```

### 4. CUAL ES EL PAIS CON MAYOR GDP PERCAPITA(PBI)

```python
df.groupby('location').mean().gdp_per_capita.sort_values(ascending=False)
```

### 5. QUE PAIS TIENE LA MAYOR CANTIDAD DE FUMADORAS FEMENINAS

```python
df.groupby('location').mean().female_smokers.sort_values(ascending=False)
```

### 6. QUE PAIS TIENE LA MAYOR CANTIDAD DE MUERTES POR MILLON

```python
df.groupby('location').mean().total_deaths_per_million.sort_values(ascending=False)
```

### 7. QUE PAÍS A TOMADO LAS MEDIDAS MAS ESTRICTAS HASTA AHORA

```python
df.groupby('location').mean().stringency_index.sort_values(ascending=False)
```

### 8. CUAL FUE EL PAIS EN ENCONTRAR 10,000 CASOS ACTIVOS POR PRIMERA VEZ

```python
df10 = df[ (df['date']>'2020-01-01') | (df['total_tests_per_thousand']>10000)]
df10
```
