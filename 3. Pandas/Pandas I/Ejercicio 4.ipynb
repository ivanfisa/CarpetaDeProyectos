### Ejercicio 4 - Investigar Como?  Seleccionar Filas y Columnas

<div class="well">Crear <strong>Selección</strong> del DataFrame con: 
   
   1.-Selección de los <strong>nombres</strong> de pasajeros <strong>mayores de 30 años.</strong>
   
   2.-Selección de los <strong>nombres</strong> de las <strong>mujeres</strong> que tienen familiares que abordaron el titanic.
    
   3.-Selección de las <strong>filas 0 a 10 y las columnas 2 a 6.</strong>
    
   
</div>

### Leemos el archivo xlsx

```python
#importamos las librerias necesarias
import pandas as pd

#le asignamos una variable al documento eliminando la primera columna para no confundir el id con la primera columna
df = pd.read_excel('data/titanic.xlsx', index_col = 0)   #index_col = 0 elimina la primera columna de la tabla  

#leemos el documento
df.head(900)
```

### Pregunta 1: seleccionar nombres de pasajeros mayores de 30 años

```python
#definimos una variable
df_mayor_a_30 = df.loc[df['Edad'] >= 30]

#mostramos los datos 
df_mayor_a_30.head(900)
```

### Pregunta 2: seleccionar nombres de mujeres que tienen familiares que aboradaron el titanic

```python
#defninimos una funcion para seleccionar solo las personas que sean mujeres (osea de sexo femenino)
mujeres= df[df['Sexo'] == 'Femenino']

#definimos una funcion para seleccionar solo a las mujeres que tengan 1 familiar o más abordo del titanic
masde1 = df[df['Familiares'] >= 1]

#juntamos las funciones anteriores
problema2 = mujeres + masde1

#seleccionamos a las mujeres que tengan 1 familiar o más abordo del titanic
problema2.head(900)
```

### Pregunta 3: seleccionar de las filas 0 a 10 y de las columna 2 a 6

```python
#incluimos nuestra dataframe pero esta vez si le permitimos el index 0
df1 = pd.read_excel('data/titanic.xlsx')

#definimos una funcion para df1 para seleccionar las filas y columnas
problema3 = df1.iloc[0:11,1:6]

#mostramos las filas de 0 a 10 y las columnas de 2 a 6
problema3
```
