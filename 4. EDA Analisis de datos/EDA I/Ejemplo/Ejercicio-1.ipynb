```python
#Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
data = pd.read_csv('datos_horarios_contaminacion_lima.csv')
data.head()
```

```python
#Verificar si hay datos nulos en el DataFrame
print('Datos nulos en el DataFrame:')

#Imprimir nulos TRUE
data.isnull()
```

```python
#Suma de datos nulos en el DataFrame
print('Suma de datos nulos en el DataFrame:')
#imprimir
data.isnull().sum()
```

```python
data.info()
```

```python
Resultad_ = pd.DataFrame(data)
```

```python
Resultad_["Fecha"] = pd.to_datetime(dict(year=Resultad_["ANO"], 
                                         month=Resultad_["MES"], 
                                         day=Resultad_["DIA"]))
Resultad_["Fecha"] 
```

```python
data.drop(['ANO', 'MES', 'DIA'], axis = 'columns', inplace=True)
```

```python
data.head()
```

```python
#grabamos en otro archivo
data.to_csv('analisis_data.csv')
```

```python
aq = pd.read_csv('analisis_data.csv', index_col = 10, parse_dates = True)
aq.head()
```

<div class="well">
<strong>- </strong>y Limpieza de datos

</div>

```python
aq.describe()
```

```python
#Reemplazando NaN en "PM 10" y "PM 2.5 mediante el promedio"

PPM_10 = aq["PM 10"].fillna(aq["PM 10"].mean())
del(aq['PM 10'])
aq.insert(4, "PM10", PPM_10)


PPM_25 = aq["PM 2.5"].fillna(aq["PM 2.5"].mean())
del(aq['PM 2.5'])
aq.insert(5, "PM 2.5", PPM_25)


PSO2 = aq["SO2"].fillna(aq["SO2"].mean())
del(aq['SO2'])
aq.insert(6, "SO2", PSO2)


PNO2 = aq["NO2"].fillna(aq["NO2"].mean())
del(aq['NO2'])
aq.insert(7, "NO2", PNO2)

PO3 = aq["O3"].fillna(aq["O3"].mean())
del(aq['O3'])
aq.insert(8, "O3", PO3)

PCO = aq["CO"].fillna(aq["CO"].mean())
del(aq['CO'])
aq.insert(9, "CO", PCO)

aq
```

```python
aq.info()
```
