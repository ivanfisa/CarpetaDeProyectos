Crear una función que represente gráficamente esta expresión:

$$\sin(2 \pi f_1 t) + \sin(2 \pi f_2 t)$$

Siendo $f_1$ y $f_2$ argumentos de entrada (por defecto $10$ y $100$) y $t \in [0, 0.5]$. Además, debe mostrar:

* leyenda,
* título "Dos frecuencias",
* eje x "Tiempo ($t$)"

y usar algún estilo de los disponibles,

La grafica resultante es:

![ejercicio5](https://user-images.githubusercontent.com/19308295/117557892-748fb200-b03d-11eb-8009-7c2ca5718ccb.png)

```python
from pylab import *
import matplotlib.pyplot as plt
import numpy as np

def f2(x):
    y = sin(x) + sin(5.0*x)
    return y

# array de valores a representar
x = linspace(0, 10*pi, 800)

p2 = plot(x, f2(x))

# Añado leyenda, tamaño de letra 10, en esquina superior derecha
legend(('Funcion 1', 'Funcion 2', 'Funcion 3'),
prop = {'size': 10}, loc='upper right')

xlabel('Tiempo (t)')
title('Dos Frecuencias')

# Creo una figura (ventana), pero indico el tamaño (x,y) en pulgadas
figure(figsize=(12, 5))

show()

```
