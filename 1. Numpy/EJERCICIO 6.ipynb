<div class="well"><strong> Grafica de la funcion 1:</strong> 
    
1. $f(x) = sin(x),  x \epsilon [-2,2]$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2,2,0.1)    #'x' va desde -2 hasta 2
y = np.array([np.sin(valor * (800/np.pi)) for valor in x])

plt.subplots(1,1,figsize=(8,8))    #aumenta de tamaño la grafica
plt.grid(True)        #agrega cuadricula
plt.plot(x, y, "--")  #cambia el color de la linea de la funcion
plt.plot(x,y)
plt.show()

#LINK DE DISEÑO: http://appgametutoriales.com/tipos-de-lineas-predeterminadas-y-personalizadas-matplotlib-python/
```

<div class="well"><strong> Grafica de la funcion 2:</strong> 
    
2. $f(x) = e^{3x},  x \epsilon [-2,4]$    Sugerencia: usar np.e para valor de 'e'.   
</div>

```python
import numpy as np
from sympy import *
x = symbols('x')
f = Function('f')

Eq(f(x), np.e**(3*x))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

x = np.arange(-2,4,0.1)
y = np.e**(3*x)

plt.subplots(1,1,figsize=(8,8))    #aumenta de tamaño la grafica
plt.grid(True)        #agrega cuadricula
plt.plot(x, y, "--")  #cambia el color de la linea de la funcion
plt.plot(x,y)
plt.show()

#LINK DE REFERENCIA: https://numython.github.io/posts/2016/02/graficar-en-python-con-matplotlib-y/
```

<div class="well"><strong> Grafica de la funcion 3:</strong> 
       
3. $f(x) = \frac{1}{x},  x \epsilon [1,10]$    

</div>

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

x = np.arange(1,10,0.1)
y = 1/x

plt.subplots(1,1,figsize=(8,8))    #aumenta de tamaño la grafica
plt.grid(True)        #agrega cuadricula
plt.plot(x, y, "--")  #cambia el color de la linea de la funcion
plt.plot(x,y)
plt.show()

#LINK DE REFERENCIA: https://numython.github.io/posts/2016/02/graficar-en-python-con-matplotlib-y/
```
